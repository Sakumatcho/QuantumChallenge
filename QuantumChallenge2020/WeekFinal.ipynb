先に3shotに絞り込んで、その中から正解になる状態を増幅するつもりでしたが、うまくいかずです、、

````python
def counter(qc, flip, auxiliary):
    for i in range(len(flip)):
        qc.mct([flip[i], auxiliary[0], auxiliary[1], auxiliary[2]], auxiliary[3], mode='noancilla')
        qc.mct([flip[i], auxiliary[0], auxiliary[1]], auxiliary[2], mode='noancilla')
        qc.ccx(flip[i], auxiliary[0], auxiliary[1])
        qc.cx(flip[i], auxiliary[0])
        qc.barrier()
        

def inv_counter(qc, flip, auxiliary):
    for i in range(len(flip)):
        qc.cx(flip[len(flip) - 1 - i], auxiliary[0])
        qc.ccx(flip[len(flip) - 1 - i], auxiliary[0], auxiliary[1])
        qc.mct([flip[len(flip) - 1 - i], auxiliary[0], auxiliary[1]], auxiliary[2], mode='noancilla')
        qc.mct([flip[len(flip) - 1 - i], auxiliary[0], auxiliary[1], auxiliary[2]], auxiliary[3], mode='noancilla')
        qc.barrier()
        
        
def init_register(qc, address, h, v):
    qc.h(address)
    qc.h(h)
    qc.h(v)

    
def post_oracle(qc, oracle_reg):
    qc.h(oracle_reg)
    qc.x(oracle_reg)
    

def encode_problem(qc, address, h, v, cluster, problem, h_ids, v_ids):
    for h_id in h_ids:
        qc.x(h[h_id])
    for v_id in v_ids:
        qc.x(v[v_id])
        
    for idx_edge, edge in enumerate(problem):
        qc.mct([
            address[0], 
            address[1], 
            address[2], 
            address[3], 
            h[edge[0]], 
            v[edge[1]], 
        ],
        cluster[idx_edge], 
        mode='noancilla'
        )

    for v_id in v_ids:
        qc.x(v[v_id])
    for h_id in h_ids:
        qc.x(h[h_id])
        
        
def inv_encode_problem(qc, address, h, v, cluster, problem, h_ids, v_ids):
    for h_id in h_ids:
        qc.x(h[h_id])
    for v_id in v_ids:
        qc.x(v[v_id])
        
    for idx_edge, edge in enumerate(problem[::-1]):
        qc.mct([
            address[0], 
            address[1], 
            address[2], 
            address[3], 
            h[edge[0]], 
            v[edge[1]], 
        ],
        cluster[len(problem) - (idx_edge + 1)], 
        mode='noancilla'
        )

    for v_id in v_ids:
        qc.x(v[v_id])
    for h_id in h_ids:
        qc.x(h[h_id])

        
def filter_problem(qc, address, idx_problem):
    bin_str = bin(idx_problem)[2:][::-1]
    ids_1 = [idx_char for idx_char, bin_char in enumerate(bin_str) if bin_char == '1'] 
    ids_0 = [idx for idx in range(4) if idx not in ids_1]
    for id_0 in ids_0:
        qc.x(address[id_0])
        

def qRAM(qc, address, h, v, cluster, problems, hv_ids):
    qc.x(h)
    qc.x(v)
    for idx_problem, problem in enumerate(problems):
        # addressの識別
        filter_problem(qc, address, idx_problem)
        
        # address -> h, v
        # mctするh, vをunique
        h_ids = hv_ids[idx_problem][0]
        v_ids = hv_ids[idx_problem][1]
        for h_id in h_ids:
            qc.mct(address, h[h_id], mode='noancilla')
        for v_id in v_ids:
            qc.mct(address, v[v_id], mode='noancilla')
        
        # address -> cluster 
        encode_problem(qc, address, h, v, cluster, problem, h_ids, v_ids)
                
        filter_problem(qc, address, idx_problem)
        qc.barrier()

        
def inv_qRAM(qc, address, h, v, cluster, problems, hv_ids):
    for idx_problem, problem in enumerate(problems[::-1]):
        # addressの識別
        filter_problem(qc, address, len(problems) - (idx_problem + 1))
        
        # address -> h, v
        # mctするh, vをunique
        h_ids = hv_ids[len(problems) - (idx_problem + 1)][0]
        v_ids = hv_ids[len(problems) - (idx_problem + 1)][1]
        # h, v -> cluster 
        inv_encode_problem(qc, address, h, v, cluster, problem, h_ids, v_ids)
        
        # h, v -> address
        # mctするh, vをunique
        for v_id in v_ids[::-1]:
            qc.mct(address, v[v_id], mode='noancilla')
        for h_id in h_ids[::-1]:
            qc.mct(address, h[h_id], mode='noancilla')
        
        filter_problem(qc, address, len(problems) - (idx_problem + 1))
        qc.barrier()
    qc.x(v)
    qc.x(h)
    
def detect_answer(qc, cluster, oracle_detect):
    qc.x(cluster)
    qc.mct(cluster, oracle_detect, mode='noancilla')
    qc.x(cluster)


def diffusion_hv(qc, h, v):
    qc.h(h)
    qc.h(v)
    qc.x(h)
    qc.x(v)
    qc.h(v[len(v) - 1])
    qc.mct(
        [
            h[0], 
            h[1], 
            h[2],
            h[3], 
            v[0], 
            v[1], 
            v[2], 
        ], 
        v[3], mode='noancilla')
    qc.h(v[len(v) - 1])
    qc.x(v)
    qc.x(h)
    qc.h(v)
    qc.h(h)


def detect3(qc, degit, oracle):
    qc.x(degit[2:])
    qc.mct(degit, oracle[0])
    qc.x(degit[2:])
    qc.barrier()
    

def detect_ansower_3(qc, cluster, degit, oracle_detect):
    qc.x(cluster)
    qc.x(degit[2:])
    qc.mct([
        cluster[0], 
        cluster[1], 
        cluster[2], 
        cluster[3], 
        cluster[4], 
        cluster[5], 
        degit[0], 
        degit[1], 
        degit[2], 
        degit[3],         
    ], oracle_detect, mode='noancilla')
    qc.x(degit[2:])
    qc.x(cluster)
    qc.barrier()
    
    
def detect_not3(qc, degit, oracle):
    true_list = [0, 1, 2, 4, 5, 6, 7, 8]
    for true_num in true_list:
        bin_str = bin(true_num)[2:][::-1]
        ids_1 = [idx_char for idx_char, bin_char in enumerate(bin_str) if bin_char == '1'] 
        ids_0 = [idx for idx in range(4) if idx not in ids_1]
        for id_0 in ids_0:
            qc.x(degit[id_0])
        qc.mct(degit, oracle[0], mode='noancilla')
        for id_0 in ids_0:
            qc.x(degit[id_0])

def diffusion_address(qc, address):
    qc.h(address)
    qc.x(address)
    qc.h(address[len(address) - 1])
    qc.mct(
        address[:-1], 
        address[-1], mode='noancilla')
    qc.h(address[len(address) - 1])
    qc.x(address)
    qc.h(address)


def diffusion_address_hv(qc, address, h, v):
    qc.h(address)
    qc.h(h)
    qc.h(v)
    qc.x(address)
    qc.x(h)
    qc.x(v)
    qc.h(v[-1])
    qc.mct(
        [
            address[0], 
            address[1], 
            address[2], 
            address[3], 
            h[0], 
            h[1], 
            h[2], 
            h[3], 
            v[0], 
            v[1], 
            v[2], 
        ], 
        v[3], mode='noancilla')
    qc.h(v[-1])
    qc.x(v)
    qc.x(h)
    qc.x(address)
    qc.h(v)
    qc.h(h)
    qc.h(address)
    

def week3_ans_func(problem_set):
    problems = np.array(problem_set).astype(int)
    hv_ids = list()
    for problem in problems:
        # mctするh, vをunique
        h_ids = np.unique(problem[:, 0])
        v_ids = np.unique(problem[:, 1])
        hv_ids.append([h_ids, v_ids])
    address = QuantumRegister(4, 'address')
    h = QuantumRegister(4, 'h')
    v = QuantumRegister(4, 'v')
    cluster = QuantumRegister(6, 'cluster')
    degit = QuantumRegister(4, 'degit')
    degit_2 = QuantumRegister(4, 'degit_2')
    oracle = QuantumRegister(1, 'oracle')
    # test
    cr = ClassicalRegister(12)
    qc = QuantumCircuit(address, h, v, cluster, degit, degit_2, oracle, cr)
    
    cnt_target = list()
    cnt_target.extend(h)
    cnt_target.extend(v)

    init_register(qc, address, h, v)
    qc.barrier()
    
    qc.x(oracle)
    qc.h(oracle)
    qc.barrier()


    # 3shotの状態に限定(まだ解を満たすかは不定)
    counter(qc, cnt_target, degit)

    detect3(qc, degit, oracle)

    inv_counter(qc, cnt_target, degit)
    
    # 振幅増幅
    diffusion_hv(qc, h, v)
    qc.h(oracle[0])
    qc.x(oracle[0])
    qc.barrier()
    
    qc.x(oracle)
    qc.h(oracle)
    qc.barrier()
#     # 問題設定
    qRAM(qc, address, h, v, cluster, problems, hv_ids)
    
    detect_answer(qc, cluster, oracle[0])
    qc.barrier()
    
    inv_qRAM(qc, address, h, v, cluster, problems, hv_ids)
        
    diffusion_hv(qc, h, v)
    
    qc.x(oracle[0])
    qc.h(oracle[0])
    qc.barrier()
    
    qc.measure(h, cr[:len(h)])
    qc.measure(v, cr[len(h):len(h) + len(v)])
    qc.measure(address, cr[len(h) + len(v): len(h) + len(v) + len(address)])
    
    qc.reverse_bits()
    return qc
````

````python
qc = week3_ans_func(problem_set)

backend = provider.get_backend('ibmq_qasm_simulator')
shots = 8192
job_exp = execute(qc, backend=backend, shots=shots)
job_monitor(job_exp, interval = 2)

results = job_exp.result()
count = results.get_counts()
sr_count = pd.Series(count)
print(sr_count.sort_values(ascending=False).iloc[:30])

qc.draw(output='mpl')
````
